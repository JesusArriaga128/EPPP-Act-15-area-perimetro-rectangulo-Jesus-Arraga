#### 📝 Redacción del Problema
#### Se requiere desarrollar un programa interactivo en Python que permita a un usuario calcular propiedades geométricas básicas de un rectángulo. El programa debe mostrar un menú persistente que permita calcular el área o el perímetro basándose en la base y la altura proporcionadas por el usuario, repitiéndose hasta que el usuario decida salir.
### 📐 Fórmulas a UtilizarPara un rectángulo con base ($b$) y altura ($h$), las fórmulas son:
1. Área: $A = b \times h$
2. Perímetro: $P = 2 \times (b + h)$
![texto del vínculo](https://raw.githubusercontent.com/AlonsoLopezGarcia128/EPPP-Act-15-area-perimetro-rectangulo-Alonso-Lopez/refs/heads/main/Area-del-Rectangulo.jpg)

```python
# ---------------------------------------------------------
# PROGRAMA: Calculadora de Rectángulos Interactiva
# ---------------------------------------------------------

def calcular_area():
    print("\n--- Cálculo de Área ---")
    base = float(input("Introduce la base del rectángulo: "))
    altura = float(input("Introduce la altura del rectángulo: "))
    area = base * altura
    print(f"✅ El área del rectángulo es: {area}")

def calcular_perimetro():
    print("\n--- Cálculo de Perímetro ---")
    base = float(input("Introduce la base del rectángulo: "))
    altura = float(input("Introduce la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"✅ El perímetro del rectángulo es: {perimetro}")

def menu():
    continuar = True

    while continuar:
        print("\n" + "="*30)
        print("     MENÚ DE GEOMETRÍA")
        print("="*30)
        print("1. Calcular Área del Rectángulo")
        print("2. Calcular Perímetro del Rectángulo")
        print("3. Salir")
        print("="*30)

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            calcular_area()
        elif opcion == "2":
            calcular_perimetro()
        elif opcion == "3":
            print("\n¡Gracias por usar el programa! Hasta pronto. 👋")
            continuar = False
        else:
            print("\n❌ Opción no válida. Por favor, intenta de nuevo.")

# Llamada principal para iniciar el programa
if __name__ == "__main__":
    menu()


print("Programa realizado por Jesus Arriaga NL = 5")
```
